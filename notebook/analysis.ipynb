Data Cleaning

```python
df1 = df.copy()
```

```python
# removing duplicate rows
df1.drop_duplicates(inplace=True)
```

```python
# Convert date columns

date_cols = ["lead_date","Reg_date","carinfo_expirydt"]
for col in date_cols:
    df1[col] = pd.to_datetime(df1[col],errors="coerce")
```

```python
# Convert premium to numeric

df1["totalpremium"] = pd.to_numeric(df1["totalpremium"],errors="coerce")
```

```python
# Remove extra spaces

obj_cols = df1.select_dtypes(include="object").columns
for col in obj_cols:
    df1[col] = df1[col].str.strip()
```

```python
# Standardizing the text

text_cols = [
    "bookedinsurer",
    "Bookedplantype",
    "utmmedium",
    "utmcampaign",
    "Make",
    "Model",
    "City",
    "State",
    "ccGroup",
    "Segment"
]

for col in text_cols:
    df1[col] = df1[col].str.title()
```

```python
# Missing values after cleaning

((df1.isna().sum()/len(df1)*100).sort_values(ascending=False))

# Missing values in bookedinsurer, Bookedplantype and totalpremium
# are valid because those leads were not converted into bookings
```

Exploratory Data Analysis

```python
# Total Leads
total_leads = df1["lead_reg"].notna().sum()

# Total Bookings
total_bookings = df1["book_reg"].notna().sum()

# Conversion Rate
conversion_rate = (total_bookings/total_leads)*100

# Total Revenue
total_revenue = df1["totalpremium"].sum()

# Average Premium
average_premium = df1["totalpremium"].mean()

# Lead Drop-off
drop_off = total_leads-total_bookings
```

```python
# KPI

print("Total Leads :",total_leads)
print("Total Bookings :",total_bookings)
print(f"Conversion Rate : {conversion_rate:.2f}%")
print("Lead Drop-off :",drop_off)
print("Total Revenue :",round(total_revenue,2))
print("Average Premium :",round(average_premium,2))
```

```python
# Top Insurers
print("Top Insurers by Revenue")
df1.groupby("bookedinsurer")["totalpremium"].sum().sort_values(ascending=False)
```

```python
# Top Plan Types
print("Top Plan Types by Revenue")
df1.groupby("Bookedplantype")["totalpremium"].sum().sort_values(ascending=False)
```

```python
# Marketing Medium
print("Bookings by UTM Medium")
df1[df1["book_reg"].notna()].groupby("utmmedium").size().sort_values(ascending=False)
```

```python
# Revenue by UTM Medium
print("Revenue by UTM Medium")
df1.groupby("utmmedium")["totalpremium"].sum().sort_values(ascending=False)
```

```python
# Campaign Performance

campaign = df1.groupby("utmcampaign").agg(
    Leads=("lead_reg","count"),
    Bookings=("book_reg",lambda x:x.notna().sum()),
    Revenue=("totalpremium","sum")
)
```

```python
campaign["Conversion %"] = round((campaign["Bookings"]/campaign["Leads"])*100,2)
campaign["Drop-off"] = campaign["Leads"]-campaign["Bookings"]

print("Top Campaign Performance")
print(campaign.sort_values("Revenue",ascending=False).head(10))
```

```python
# Vehicle Age
print("Bookings by Vehicle Age")
df1[df1["book_reg"].notna()].groupby("Vehicle_Age").size().sort_values(ascending=False)
```

```python
# Vehicle Made
print("Top Vehicle Makes")
df1[df1["book_reg"].notna()].groupby("Make").size().sort_values(ascending=False).head(10)
```

```python
# Vehicle Model
print("Top Vehicle Models")
df1[df1["book_reg"].notna()].groupby("Model").size().sort_values(ascending=False).head(10)
```

```python
# City Analysis
print("Top Cities")
df1[df1["book_reg"].notna()].groupby("City").size().sort_values(ascending=False).head(10)
```

```python
# CC Group Purchase Rate
cc_group = df1.groupby("ccGroup").agg(
    Leads=("lead_reg","count"),
    Bookings=("book_reg",lambda x:x.notna().sum())
)

cc_group["Purchase Rate %"] = round((cc_group["Bookings"]/cc_group["Leads"])*100,2)

print("CC Group Purchase Rate")
print(cc_group.sort_values("Purchase Rate %",ascending=False))
```
